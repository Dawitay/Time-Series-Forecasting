```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define ticker symbols
tickers = ['TSLA', 'BND', 'SPY']

# Define time period
start_date = '2015-07-01'
end_date = '2025-07-31'

# Download data for each ticker
data_tsla = yf.download('TSLA', start=start_date, end=end_date)
data_bnd = yf.download('BND', start=start_date, end=end_date)
data_spy = yf.download('SPY', start=start_date, end=end_date)
```

```python
print("TSLA Data:")
print(data_tsla.head())

print("\nBND Data:")
print(data_bnd.head())

print("\nSPY Data:")
print(data_spy.head())
```

```python
data_tsla['Asset'] = 'TSLA'
data_bnd['Asset'] = 'BND'
data_spy['Asset'] = 'SPY'
```

```python
# Combine all datasets
data_all = pd.concat([data_tsla, data_bnd, data_spy])
data_all.reset_index(inplace=True)  # Reset index to make 'Date' a column
```

```python
data_all.to_csv('../src/gmf_assets_data.csv', index=False)
```

```python
import pandas as pd

# Load your original CSV (adjust the filename)
df = pd.read_csv('../src/gmf_assets_data.csv')

# Inspect the columns
print("Columns in original CSV:")
print(df.columns)

# Since your CSV has repeated groups of columns per asset (TSLA, BND, SPY),
# first rename columns to unique names:

# Manually rename columns based on your CSV structure,
# adjust these lists if your column order is different

tsla_cols = ['Date', 'TSLA_Open', 'TSLA_High', 'TSLA_Low', 'TSLA_Close', 'TSLA_Volume']
bnd_cols = ['Date', 'BND_Open', 'BND_High', 'BND_Low', 'BND_Close', 'BND_Volume']
spy_cols = ['Date', 'SPY_Open', 'SPY_High', 'SPY_Low', 'SPY_Close', 'SPY_Volume']

# Your original df probably looks like this:
# Columns: ['Date', 'Close', 'High', 'Low', 'Open', 'Volume', 'Asset', 'Close', 'High', 'Low', 'Open', 'Volume', 'Close', 'High', 'Low', 'Open', 'Volume']
# So rename accordingly to avoid duplicates:
df.columns = [
    'Date',
    'TSLA_Close', 'TSLA_High', 'TSLA_Low', 'TSLA_Open', 'TSLA_Volume',
    'Asset',  # this column seems to have asset names but not useful in this format
    'BND_Close', 'BND_High', 'BND_Low', 'BND_Open', 'BND_Volume',
    'SPY_Close', 'SPY_High', 'SPY_Low', 'SPY_Open', 'SPY_Volume'
]

# Drop the 'Asset' column as itâ€™s not needed
df = df.drop(columns=['Asset'])

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Now create separate DataFrames per asset, dropping rows where Close is NaN
tsla = df[['Date', 'TSLA_Open', 'TSLA_High', 'TSLA_Low', 'TSLA_Close', 'TSLA_Volume']].dropna(subset=['TSLA_Close']).copy()
tsla['Asset'] = 'TSLA'
tsla.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Asset']

bnd = df[['Date', 'BND_Open', 'BND_High', 'BND_Low', 'BND_Close', 'BND_Volume']].dropna(subset=['BND_Close']).copy()
bnd['Asset'] = 'BND'
bnd.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Asset']

spy = df[['Date', 'SPY_Open', 'SPY_High', 'SPY_Low', 'SPY_Close', 'SPY_Volume']].dropna(subset=['SPY_Close']).copy()
spy['Asset'] = 'SPY'
spy.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Asset']

# Concatenate into one long DataFrame
long_df = pd.concat([tsla, bnd, spy], ignore_index=True)

# Sort by Date and Asset for neatness
long_df = long_df.sort_values(by=['Date', 'Asset']).reset_index(drop=True)

# Save to new CSV
long_df.to_csv('../src/gmf_assets.csv', index=False)

print(long_df.head())
```
